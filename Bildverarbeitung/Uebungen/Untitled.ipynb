```python
imgName = '../../images/DonaldDuck01.tiff'
i = cv2.imread(imgName, cv2.IMREAD_COLOR)
Icolor = cv2.cvtColor(i, cv2.COLOR_BGR2RGB )


plt.imshow(Icolor)
```

```python
ir = Icolor[:,:,0]
ig = Icolor[:,:,1]
ib = Icolor[:,:,2]

plt.subplot(2,2,1); plt.imshow(ir,cmap = 'Reds')
plt.subplot(2,2,2); plt.imshow(ig,cmap = 'Blues')
plt.subplot(2,2,3); plt.imshow(ib,cmap = 'Greens')
```
