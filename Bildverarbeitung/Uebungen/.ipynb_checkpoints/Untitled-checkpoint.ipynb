```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Start

```python
imgName = '../images/DonaldDuck01.tiff'
igray = cv2.imread(imgName, cv2.IMREAD_GRAYSCALE)

plt.imshow(igray, cmap = 'gray')
```

```python
imgName = '../images/DonaldDuck01.tiff'
i = cv2.imread(imgName, cv2.IMREAD_COLOR)
Icolor = cv2.cvtColor(i, cv2.COLOR_BGR2RGB )

plt.imshow(Icolor)
```

```python
plt.figure(figsize=(15, 15))

ir = Icolor[:,:,0]
ig = Icolor[:,:,1]
ib = Icolor[:,:,2]

plt.subplot(1,3,1); plt.imshow(ir,cmap = 'gray')
plt.subplot(1,3,2); plt.imshow(ig,cmap = 'gray')
plt.subplot(1,3,3); plt.imshow(ib,cmap = 'gray')
```

```python
plt.figure(figsize=(15, 15))

irr = i.copy()
irr[:,:,1] = 0
irr[:,:,2] = 0

igg = i.copy()
igg[:,:,0] = 0
igg[:,:,2] = 0

ibb = i.copy()
ibb[:,:,0] = 0
ibb[:,:,1] = 0

plt.subplot(1,3,1); plt.imshow(irr)
plt.subplot(1,3,2); plt.imshow(igg)
plt.subplot(1,3,3); plt.imshow(ibb)
```

```python
plt.figure(figsize=(15, 15))

irg = i.copy()
irg[:,:,2] = 0



ibg = i.copy()
ibg[:,:,0] = 0


plt.subplot(1,3,1); plt.imshow(ibg)
plt.subplot(1,3,2); plt.imshow(irg)
```
