```python
import numpy as np
import math

import matplotlib.pyplot as plt
import seaborn as sns

from course_utils import *
```

```python
sns.set()

save_graph(True)
graph_dir = '../../../Abbildungen/'
```

```python
def clamp(t): 
    if (t<0): return 0
    elif (t>255):
        return 255
    else: return t


def clamp01(t): 
    if (t<0): return 0
    elif (t>1):
        return 1
    else: return t
    
def Ident(t): 
    return t

def Invers(t): 
    return 255-t

def Log(t): 
    if (t==0): return 0
    else: return math.log(t)*255.0/(math.log(255))
    
def Wurzel(t): 
    return math.sqrt(t)*math.sqrt(255)

def Quadrat(t): 
    return t*t

def NegMul(t): 
    return 1 - (1-t)*(1-t)

def FarbNach(t): 
    if (t == 0): return 0 
    else: return clamp01(1-(1-t)/t)
    
def LinNach(t): 
    return clamp01(2*t-1)

def FarbAb(t): 
    if (t != 1): return clamp01(t/(1-t))
    else: return 1

def LinAb(t): 
    return clamp01(2*t)

def Ineinander(t): 
    if (t<=0.5): return 2*t*t
    else: return 1-2*(1-t)*(1-t)
    
def WeichesLicht(t): 
    if (t<=0.5): return (2*t-1)*(t-t*t)+t
    else: return (2*t-1)*(sqrt(t)-t)+t
    
def HartesLicht(t): 
    if (t<=0.5): return 2*t*t
    else: return 1-2*(1-t)*(1-t)

def StrahlendesLicht(t): 
    if (t<=0.5 and t != 0): 
        return clamp01(1-(1-t)/(2*t))
    elif (t != 1):
        return clamp01(t/(2*(1-t)))
    elif(t==0): return 0
    else: return 1

def LinearesLicht(t): 
    return clamp01(3*t-1)

def Exp(a, t): 
    return 255.0/(np.exp(a*255.0)-1)*(np.exp(a*t)-1)

def Kontrast(t): 
    return 1.5*t

def Bright(t): 
    return t+20

def Schwelle(t): 
    if (t<120): return 0
    else: return 255

    
def _sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
        
def Sigmoid(t):
    t = (t/255*6-3)
    g = _sigmoid(t)-_sigmoid(-3)            
    return (np.uint8)(g*255/(_sigmoid(3)-_sigmoid(-3)))

def apply_LUT(f):
    G = np.linspace(0,255,256,dtype=np.uint8)
    res = np.ndarray((256), dtype=np.uint8)
    for g in G:
        res[g] = clamp(f(g))
    return G, res
 

def display_LUT_OLD(f, step=None):
    G, lut = apply_LUT(f)
    
    plt.figure(figsize=(5,5))
    plt.plot((0,255), (0,255), c='gray', linestyle='dashed', linewidth=1)
    
    if step is None:
        plt.plot(G, lut, c='b')
    else:
        plt.plot(G[0:step], lut[0:step], c='b')
        plt.plot(G[step:], lut[step:], c='b')

    plt.xlabel('$g$')
    plt.ylabel('$g^{\prime}$')
    #plt.xticks([0, 255])
    #plt.yticks([0, 255])

    
def display_func(x, y, step=None, color='b', label=None):
    
    if step is None:
        plt.plot(x, y, c=color, label=label)
    else:
        plt.plot(x[0:step], y[0:step], c=color, label=label)
        plt.plot(x[step:], y[step:], c=color, label=label)


def display_LUT(f, step=None):
    G, lut = apply_LUT(f)
    
    plt.figure(figsize=(5,5))
    plt.plot((0,255), (0,255), c='gray', linestyle='dashed', linewidth=1)
    display_func(G, lut, step)

    plt.xlabel('$g$')
    plt.ylabel('$g^{\prime}$')
    #plt.xticks([0, 255])
    #plt.yticks([0, 255])



    
```

```python
display_LUT(Bright)
show_graph(graph_dir+'PO_brightness.pdf')
```
