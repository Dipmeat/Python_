### ... let's calculate the aritmetic mean, best estimate or best guess ...

$$\bar x  =\frac{1}{N}\sum_{i=1}^N x_i$$

... mit ...

  $$\triangle x_i = x_i-\bar x$$

... ergibt sich für die Varianz ...

$$s^2  =\frac{1}{N-1}\sum_{i=1}^N (\triangle x_i)^2$$

... und für die Standabweichung  

$$s=\sigma=\sqrt{\frac{1}{N-1}\sum_{i=1}^N (\triangle x_i)^2}$$

... als mittlerer Fehler einer einzelnen Messung

Für den Standardfehler gilt:

$$s_\bar x=\sigma_\bar x= \frac{\sigma}{\sqrt{N}} = \sqrt{\frac{1}{N(N-1)}\sum_{i=1}^N (\triangle x_i)^2}$$ dsdsd

### Es folgt für das obige Beispiel mit 100 Werten:

```python
from IPython.display import display, Markdown
display(Markdown(
   rf"""
$\bar x$ = ${np.mean(randomnoise)}$
"""))
display(Markdown(
   rf"""
$s^2$ = ${np.var(randomnoise)}$
"""))
display(Markdown(
   rf"""
$s =\sigma$ = ${np.std(randomnoise)}$
"""))
display(Markdown(
   rf"""
$\sigma_\bar x$ = ${np.std(randomnoise, ddof=1)/np.sqrt(np.size(randomnoise))}$
"""))
```

### Abbildung in die 2. Dimension

```python
import numpy as np    
nhist = 100
number_of_values = 1000
matrix = np.zeros((2*nhist+50,nhist)) 
mu = 0.0
sigma = 0.1
randomnoise = np.random.normal(mu, sigma, number_of_values)





```

```python
nph =np.array(list(np.histogram(randomnoise,bins=100)))

```

```python
for i in range(1,nhist):

    randomnoise = np.random.normal(mu, sigma, number_of_values)+i

    nph =np.array(list(np.histogram(randomnoise,bins=100)))

    
    for j in range(1,nhist):
        matrix[j+i,i] = nph[0][j]
        
        
```

```python
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111)
ax.set_title('Histogram ColorMap')
plt.imshow(matrix)


cax = fig.add_axes([0.12, 0.1, 0.78, 0.8])
cax.get_xaxis().set_visible(False)
cax.get_yaxis().set_visible(False)
cax.patch.set_alpha(0)
cax.set_frame_on(False)
plt.colorbar(orientation='vertical')
plt.show()
```
